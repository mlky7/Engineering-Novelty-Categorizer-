```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\Dhamu\Downloads\df.csv")
```

```python
print("Dataset Shape:",df.shape)
print("\nColumns:",df.columns)
print("\nFirst 5 Rows:")
print(df.head())

print("\nClass Distribution:")
print(df["novelty tier"].value_counts())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="novelty tier", data=df)

plt.title("Distribution of Novelty Tiers")
plt.xlabel("Novelty Tier")
plt.ylabel("Count")
plt.show()
```

```python
df["text_length"] = df["clean_text"].apply(lambda x: len(str(x).split()))
print("\nText Length Statistics:")
print(df["text_length"].describe())

plt.figure(figsize=(8,5))
plt.hist(df["text_length"], bins=20)

plt.title("Distribution of Text Length (Word Count)")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")
plt.show()
```

```python
avg_length = df.groupby("novelty tier")["text_length"].mean()
plt.figure(figsize=(6,4))
avg_length.plot(kind="bar")
plt.title("Average Text Length per Novelty Tier")
plt.xlabel("Novelty Tier")
plt.ylabel("Average Word Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="novelty tier", y="text_length", data=df)
plt.title("Text Length Variation by Novelty Tier")
plt.xlabel("Novelty Tier")
plt.ylabel("Word Count")
plt.show()
```
